## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genders = pd.DataFrame(purchase_data.loc[:,['SN','Gender']])
unique_genders = genders.drop_duplicates(subset = 'SN')

unique_genders = pd.DataFrame(unique_genders['Gender'].value_counts())
total_people = unique_genders['Gender'].sum()
unique_genders["Percentage of Players"] = [round(gender/total_people *100 ,2) for gender in unique_genders['Gender']]

unique_genders.head()
# print(unique_genders['Gender'])
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
male_data = purchase_data.loc[purchase_data['Gender'] == 'Male',:]
female_data = purchase_data.loc[purchase_data['Gender'] == 'Female',:]
other_data = purchase_data.loc[purchase_data['Gender'] == 'Other / Non-Disclosed',:]

# male_data.head()
# female_data.head()
# other_data.head()


purchases_gender = purchase_data.loc[:,['Gender']]
purchases_gender = pd.DataFrame(purchases_gender['Gender'].value_counts())
purchases_gender = purchases_gender.rename(columns={"Gender": "Purchase Count"})

purchases_gender["Average Purchase Price"] = [male_data['Price'].mean(), 
                                              female_data['Price'].mean(),
                                              other_data['Price'].mean()]



purchases_gender["Total Purchase Value"] = [male_data['Price'].sum(), 
                                              female_data['Price'].sum(),
                                              other_data['Price'].sum()]



purchases_gender["Avg Total Purchase per Person"] = [round(a/b,2) for a, b in  zip(purchases_gender["Total Purchase Value"], unique_genders['Gender'])]

purchases_gender["Average Purchase Price"] = [round(x,2) for x in purchases_gender["Average Purchase Price"]]
purchases_gender.head()



```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

ages = pd.DataFrame(purchase_data.loc[:,['SN','Age']])
ages = ages.drop_duplicates(subset = 'SN')
del ages['SN']

ages["Ages Summary"] = pd.cut(ages['Age'], age_bins, labels = group_names)


ages_data = pd.DataFrame(ages["Ages Summary"].value_counts().sort_index())

ages_data = ages_data.rename(columns = {"Ages Summary": "Total Count"})

ages_data["Percentage of Players"] = [round(count/len(ages['Age'])*100,2) for count in ages_data["Total Count"]]

ages_data
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_analysis = purchase_data

purchase_analysis["Ranges"] = pd.cut(purchase_analysis['Age'], age_bins, labels = group_names)

age_range_data = purchase_analysis.groupby(['Ranges'])


purchase_count = age_range_data['Purchase ID'].count()

total_purchase_value = age_range_data['Price'].sum()

unique_players = purchase_analysis.drop_duplicates(subset = 'SN').groupby(['Ranges'])
unique_players = unique_players['SN'].count()



purchase_by_age = pd.DataFrame(age_range_data['Purchase ID'].count())

purchase_by_age = purchase_by_age.rename(columns = {'Purchase ID': 'Purchase \nCount'})

purchase_by_age['Average Purchase Price'] = round((total_purchase_value/purchase_count),2)

purchase_by_age['Total Purchase Value'] = total_purchase_value

purchase_by_age['Avg Total Purchase Per Person'] = round((total_purchase_value/unique_players),2)

purchase_by_age.head()




```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data[['SN','Price']]

top_spenders_df = top_spenders.groupby('SN').count().sort_values(by = ['Price'], ascending = False)

top_spend_total = purchase_data.groupby('SN')['Price'].sum()

top_spend_count = top_spenders_df['Price']

top_spend_avg = round(top_spend_total/top_spend_count,2)

top_spenders_df['Average Purchase Price'] = top_spend_avg
top_spenders_df['Total Purchase Value'] = top_spend_total

top_spenders_df = top_spenders_df.rename(columns = {'Price': 'Purchase Count'})

top_spenders_df


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items = purchase_data[['Item ID','Item Name','Price']]

popular_items = pop_items.groupby(['Item ID', 'Item Name'])
popular_items = popular_items.count().sort_values(by = 'Price', ascending = False)

total_purchase_value = pop_items.groupby(['Item ID', 'Item Name'])['Price'].sum()
purchase_count = pop_items.groupby('Item Name').count()

popular_items['Item Price'] = total_purchase_value/popular_items['Price']
popular_items['Total Purchase Value'] = total_purchase_value

popular_items.head()



```
